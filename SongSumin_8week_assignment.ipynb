* 동아리 회원들의 정보를 저장하는 Class를 만들 것이다.
* Member 클래스의 코드는 수정하지 않는다.
* Python 기본 Data Collection Type인 List를 사용하지 않는다.
* 구현해야할 클래스인 Members의 명세는 아래와 같다
* Linked List를 사용해서 풀이해야 함, Linked List에 대한 설명은 이 블럭의 마지막 부분 [참고 자료] 에 있음

### ```Class Member```

def __init__(self, id, name, year, gender, registered):

### ```Class Members``` 
##### 멤버 변수
* ```head = Member('headData', 'headData', 'headData', 'headData', False)```
* ```number = 0```

##### 멤버 함수
* ```addMember(self, id, year, name, gender, registered)```
    * id: 학번 (str)
    * year: 가입년도 (str)
    * name: 이름 (str)
    * gender: 성별 (str)
    * registered: 회비 납부 여부 (boolean)
    <br><br>
* ```delMember(self, id)```
    * 입력 받은 학번과 일치하는 회원을 삭제한다
    * 일치하는 회원이 없거나 회원 수가 0 이라면 "can not find %s member" %id를 출력하고 함수를 종료한다
<br><br>
* ```editMember(self, id)```
    * 입력 받은 학번과 일치하는 회원의 정보를 수정하는 함수를 불러온다
    * 일치하는 회원이 없거나 회원 수가 0 이라면 "can not find %s member" %id를 출력하고 함수를 종료한다
<br><br>
* ```editMemberDetail(self)```
    * 회원의 정보를 출력하고, 다음 줄에 어떤 정보를 수정할지 물어보는 텍스트를 출력한 후 사용자로 부터 입력 받는다.
    * 대체할 정보를 입력 받은 후 저장한다
<br><br>
* ```isLen(self)```
    * 동아리의 회원 수를 반환한다
<br><br>
* ```newSeason(self)```
    * 새 학기를 맞아 회비 납부 여부를 초기화한다
    * 모든 회원의 회비 납부 여부를 False로 바꾼다
<br><br>
* ```infoMem(self, id)```
    * 입력 받은 학번과 일치하는 회원의 정보를 출력한다
<br><br>
* ```printMembers(self)```
    * 모든 회원의 정보를 아래와 같이 보기 좋게 출력한다

|id|year|name|gender|registered|
|:-:|:-:|:-:|:-:|:-:|
|2017103000|2017|홍길동|남|1|


# 참고 자료
[강의노트 18. 자료구조 - LinkedList (링크드 리스트)](https://wayhome25.github.io/cs/2017/04/17/cs-19/) - 초보몽키의 개발공부로그

````python
class Member:
    id = None
    name = None
    year = None
    gender = None
    registered = None
    next = None

    def __init__(self, id, year, name, gender, registered):
        self.id = id
        self.name = name
        self.year = year
        self.gender = gender
        self.registered = registered
    def getInfo(self):
        return self.id, self.year, self.name, self.gender, self.registered
````

````python
class Members:
    head = Member('headData', 'headData', 'headData', 'headData', False)
    number = 0
    
    def addMember(self, id, year, name, gender, registered):

        newMember = Member(id, year, name, gender, registered)

        location = self.head

        while location.next != None:
            location = location.next

        location.next = newMember

        self.number += 1


    def delMember(self, id):

        if self.isLen() == 0 :
            print("can not find %s member" % id)
            return "can not find %s member" % id

        location = self.head
        prev = None

        moreToSearch = True

        while moreToSearch:
            if location.next == None:
                print("can not find %s member" % id)
                return "can not find %s member" % id

            if not location.next.id == id:
                location = location.next
                continue

            prev = location
            location = location.next
            moreToSearch = False

        prev.next = location.next
        del location
        self.number -= 1


    def editMember(self, id):

        if self.isLen() == 0:
            print("can not find %s member" % id)
            return "can not find %s member" % id

        location = self.head

        moreToSearch = True

        while moreToSearch:
            if location.next == None:
                print("can not find %s member" % id)
                return "can not find %s member" % id

            if location.next.id == id:
                self.editMemberDetail(location.next)

                return

            location = location.next
      

    def editMemberDetail(self, editMember):
        self.printMembers()
        info = input("어떤 정보를 바꿀래요: ")
        if info == "id":
            new_id = input("수정할 id 입력: ")
            editMember.id = new_id

        if info == "year":
            new_year = input("수정할 year 입력: ")
            editMember.year = new_year
    
        if info == "name":
            new_name = input("수정할 이름 입력: ")
            editMember.name = new_name

        if info == "gender":
            new_gender = input("수정할 gender 입력: ")
            editMember.gender = new_gender

        if info == "registered":
            new_registered = input("수정할 registered 입력: ")
            editMember.registered = new_registered

    def isLen(self):
        return self.number

    def newSeason(self):
        location = self.head

        while location.next != None:
            location.next.registered == False
            location = location.next

    def infoMem(self, id):
        

        location = self.head

        moreToSearch = True

        while moreToSearch:
            if location.next == None:
                return
       

            if location.next.id == id:
                print("|{:^12}|{:^12}|{:^12}|{:^12}|{:^12}|".format("id","name","year","gender","registered"))
                print("|{:^12}|{:^9}|{:^12}|{:^11}|{:^12}|".format(location.next.id,location.next.name,location.next.year,location.next.gender,location.next.registered))
                return
            location = location.next

    def printMembers(self):

        location = self.head

        while location.next != None:
            print("|{:^12}|{:^12}|{:^12}|{:^12}|{:^12}|".format("id","name","year","gender","registered"))
            print("|{:^12}|{:^9}|{:^12}|{:^11}|{:^12}|".format(location.next.id,location.next.name,location.next.year,location.next.gender,location.next.registered))
            location = location.next
````

## 아래의 테스트 코드가 문제 없이 동작하여야 함

````python
tgwing = Members()

tgwing.addMember('2017103000', '2017', '홍길동', '남', True)
tgwing.addMember('2018104000', '2018', '길동홍', '남', True)
tgwing.addMember('2019105000', '2019', '동홍길', '남', True)
tgwing.printMembers()
print("현재 회원 수:", tgwing.isLen())
print()

tgwing.delMember('2018104000')
tgwing.delMember('2018104000')
tgwing.printMembers()
print("현재 회원 수:", tgwing.isLen())
print()

tgwing.editMember('2019105000')
print()

tgwing.printMembers()
print()

tgwing.infoMem('2017103000')
tgwing.newSeason()
tgwing.printMembers()
````
